# Graph plot

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['Feature_1'],data['Feature_2'] )
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.title('Scatter plot of the Features')
plt
```

```python
from sklearn import cluster
```

```python
k_means=cluster.KMeans(n_clusters=11, max_iter=100, random_state=1)
```

```python
k_means
```

```python
k_means.fit(data)
```

```python
data['Cluster'] = k_means.predict(data)
data.head()
```

# The Cluster Plot using kmeans feature.

```python
plt.scatter(data['Feature_1'],data['Feature_2'], c=data['Cluster'], cmap='winter')
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.title('Scatter plot of Feature 1 and Feature 2')
plt
```
